1. WAPP to draw line diagram(liner graph)

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4,5]
y = [2,3,5,7,11]

plt.plot(x, y,marker='s') 
plt.title('Line diagram') 
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()  
```

2. WAPP to draw multi line diagram(Line graph)

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4,5]
y = [2,3,5,7,11]
y1=[5,4,6,10,13]
y2=[6,8,9,12,7]

plt.plot(x, y,marker='s') 
plt.plot(x, y1,marker='o') 
plt.plot(x, y2,marker='d') 
plt.title('Line diagram') 
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()  
```

3. WAPP to draw an area diagram (Area plot)

```python
import plotly.express as px

x = [1, 2, 3, 4,5]
y = [2,3,5,7,11]

fig = px.area(x=x, y=y)

fig.update_layout(title='Area Diagram',yaxis_title='Y-axis',xaxis_title='X-axis')

fig.show()
```

4. WAPP to draw a multiple area diagram(Area plot)

```python
import plotly.express as px

x = [1, 2, 3, 4,5]
y = [2,3,5,7,11]
y1=[5,4,6,10,13]
y2=[6,8,9,12,7]

fig = px.area(x=x, y=[y,y1,y2],labels={'y':'y-axis'})

fig.update_layout(title='Area Diagram',xaxis_title='X-axis')

fig.show()
```
