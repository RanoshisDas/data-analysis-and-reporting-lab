1. Write a python program to generate word cloud and find most important words and find frequency.

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import re

# Sample text
text = """
Hello i am Ranoshis Das. I am a student in B.Tech CSE(data science).This is a code for word cloud, where we going 
see any word frequency in this text paragraph.
"""
# text="""
# Your sample text goes here. Replace this with the text you want to analyze.
# """

text = re.sub(r'[^\w\s]', '', text).upper()

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# Find word frequencies
words = text.split()
word_counts = Counter(words)

# Main function
print("Most common words and their frequencies:")
for word, freq in word_counts.most_common(10):
    print(f"{word}: {freq}")

#Your sample text goes here. Replace this with the text you want to analyze.
```
