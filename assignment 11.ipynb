1. Write a python program to generate word cloud and find most important words and find frequency.

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import re

# Sample text
text = """
Hello i am Ranoshis Das. I am a student in B.Tech CSE(data science).This is a code for word cloud, where we going 
see any word frequency in this text paragraph.
"""

# Preprocess the text
def preprocess_text(text):
    # Remove punctuation and make lowercase
    text = re.sub(r'[^\w\s]', '', text).upper()
    return text

# Generate word cloud
def generate_word_cloud(text):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Find word frequencies
def find_word_frequencies(text):
    words = text.split()
    word_counts = Counter(words)
    return word_counts

# Main function

processed_text = preprocess_text(text)
generate_word_cloud(processed_text)
word_frequencies = find_word_frequencies(processed_text)
    
# Print the most common words and their frequencies
print("Most common words and their frequencies:")
for word, freq in word_frequencies.most_common(10):
    print(f"{word}: {freq}")

#Your sample text goes here. Replace this with the text you want to analyze.
```
