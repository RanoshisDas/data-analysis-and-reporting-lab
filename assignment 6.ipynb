```python
import matplotlib.pyplot as plot

x=["CSE",'AI','DS','Core','CS']
y=[2,3,4,6,8]

plot.ylabel('department')
plot.xlabel('values')
plot.barh(x,y)
plot.show()
```

3. Write a python program to implement a double bar graph

```python
import matplotlib.pyplot as plt 
import numpy as np
X = ['Group A','Group B','Group C','Group D'] 
Ygirls = [10,20,20,40] 
Zboys = [20,30,25,30] 
X_axis = np.arange(len(X)) 
  
plt.bar(X_axis - 0.2, Ygirls, 0.4, label = 'Girls') 
plt.bar(X_axis + 0.2, Zboys, 0.4, label = 'Boys') 
  
plt.xticks(X_axis, X) 
plt.xlabel("Groups") 
plt.ylabel("Number of Students") 
plt.title("Number of Students in each group") 
plt.legend() 
plt.show() 
```

5. Write a python program to implement a stack bar graph

```python
import matplotlib.pyplot as plt

x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 10, 30]
y2 = [20, 25, 15, 25]
 
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.show()
```

6. Write a python program to implement a line bar graph

```python
import matplotlib.pyplot as plot

x=["CSE",'AI','DS','Core','CS']
y=[5,3,4,6,3]

plot.bar(x,y,color='g')
plot.plot(x,y,marker='o',color='black')
plot.xlabel('department')
plot.ylabel('values')
plot.show()
```
