1. bar chart from csv.

```python
import pandas as pd
import matplotlib.pyplot as pt

cv=pd.read_csv("https://raw.githubusercontent.com/RanoshisDas/data-analysis-and-reporting-lab/refs/heads/main/data.csv")
h=cv["Height"]
w=cv["Weight"]

pt.bar(h,w)
pt.show()
```
